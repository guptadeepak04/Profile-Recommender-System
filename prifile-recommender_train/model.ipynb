```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Assuming your dataset is already cleaned and loaded into 'profile_data'

# Select features for the model (adjust based on relevant columns)
features = ['Graduation Year', 'Years of Experience']  # Select numerical columns

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data[features])

# Initialize the KNN model
knn = NearestNeighbors(n_neighbors=5, algorithm='auto', n_jobs=1)  # Limit to 1 thread
knn.fit(X_scaled)

# Function to recommend alumni by university name
def recommend_alumni_by_university(university_name):
    # Filter the dataset for the specified university
    university_data = data[data['University'] == university_name]
    
    if university_data.empty:
        print(f"No alumni found from {university_name}")
        return []

    # Get the features for alumni from that university
    university_features = university_data[features]
    
    # Scale the features
    university_features_scaled = scaler.transform(university_features)

    # Find the nearest neighbors for the university alumni
    distances, indices = knn.kneighbors(university_features_scaled, n_neighbors=5)

    # Get the indices of recommended alumni
    recommended_alumni = data.iloc[indices.flatten()]

    return recommended_alumni[['Name of Student', 'University', 'Job Title', 'Company', 'Technical and Non-technical Skills']]

# Example usage
university_name = 'MIT'
recommended_alumni = recommend_alumni_by_university(university_name)
print("Recommended Alumni from the same university:")
print(recommended_alumni)
```

```python
# Example usage
university_name = 'Stanford University'  # Replace with the university you want to search for
recommended_alumni = recommend_alumni_by_university(university_name)

print("Recommended Alumni from the same university:")
print(recommended_alumni)
```
